```python
import numpy as np
import matplotlib.pyplot as plt

# CSVファイルを読み込む
# （年齢、性別（1:男性、2:女性）、身長（cm）、体重（kg））
data = np.loadtxt('hobbie2018ex.csv', delimiter=',')

# 条件を設定する
data = data[data[:,0]>=1,:] # 1歳以上
data = data[data[:,0]<=80,:] # 80歳以下
data = data[data[:,1]==2,:] # 女性

# 身長と体重をxとyとする
x = data[:, 2]
y = data[:, 3]

# 平均を計算する
x_mean = np.mean(x)
y_mean = np.mean(y)

# 相関係数を計算する
correlation = np.corrcoef(x, y)[0, 1]

# 計算結果を出力する
print('X mean:', x_mean)
print('Y mean:', y_mean)
print('Correlation coefficient:', correlation)

# 散布図をプロットする
plt.scatter(x, y)
plt.xlabel('X (height)')
plt.ylabel('Y (weight)')
plt.show()
```

import cv2 # コンピュータビジョンライブラリcv2をロード

roi_y1 = 350

# ROI表示
