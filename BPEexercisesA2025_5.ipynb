````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd # データ解析支援ライブラリpandasをインポート

filename = '/content/BIG5/data.csv'

# CSVファイルの読み込み
df = pd.read_csv(filename, delimiter='\t')

# 被験者属性を取得
race = df['race']
age = df['age']
gender = df['gender']

# 回答カラム（E1からO10）を抽出してNumPy配列に変換
selected_columns = df.columns[7:57]# E1からO10は7番目から56番目のカラム
data_all = df[selected_columns].to_numpy()

# 回答ラベルのリストを取得（7-56列目）
labels = df.columns[7:57]

# 被験者のうちの約半分を無作為抽出
rand_array = np.random.choice([0, 1], size=data_all.shape[0])
data = data_all[rand_array == 0, :]

# 相関行列の計算
dist_matrix = np.corrcoef(data.T)

# 相関行列の表示
plt.figure(figsize=(10, 10))
plt.imshow(dist_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.xlabel('Questions')
plt.ylabel('Questions')
plt.yticks(range(len(labels)), labels)
plt.colorbar()
plt.title('Correlation Matrix')
plt.show()
````

集団サンプルの相関行列を特徴とした主成分分析(PCA)
```python
from sklearn.decomposition import PCA #機械学習ライブラリsklearnから主成分分析PCAをインポート

# データ保持用の空リストを準備する
matrix_array = []
label_array = []

# 各集団の相関行列を求める
for age_group in [1,2,3,4]:
  data = data_all[(gender==1) & (age>=age_group*10) & (age<age_group*10+10),:]
  dist_matrix = np.corrcoef(data.T)
  vector = dist_matrix[upper_tri_indices] # 相関行列を1次元特徴ベクトル化
  matrix_array.append(vector) # 特徴ベクトル化をリストに追加
  label_array.append(f'M{age_group*10}s') # ラベルをつけてリストに追加（例：M20 男性20代）

# 各集団のデータを行列形式にまとめる
data_array = np.vstack(matrix_array)
labels = np.hstack(label_array)

# PCAによる次元削減
pca = PCA(n_components=2)
pca_results = pca.fit_transform(data_array)

# 散布図をプロット
plt.figure(figsize=(12, 8))
plt.scatter(pca_results[:, 0], pca_results[:, 1])

# 各ポイントにラベルを追加
for i, label in enumerate(labels):
    plt.annotate(label, (pca_results[i, 0], pca_results[i, 1]))

plt.title('PCA 2D Projection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
