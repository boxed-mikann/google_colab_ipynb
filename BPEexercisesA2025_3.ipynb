```python
### 動画が正しく読み込めることを確認し、ROIを設定する

import cv2
import matplotlib.pyplot as plt
import numpy as np

# 動画ファイルの読み込み
video_path = 'output.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f'ファイル {video_path} が読み込めませんでした。')
else:
    print(f'ファイル {video_path} は正しく読み込まれました。')

if cap.isOpened():
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    print(f'総フレーム数: {total_frames}')
    print(f'幅: {width}')
    print(f'高さ: {height}')
    print(f'FPS: {fps}') # frames per second

# 矩形ROIの指定（左上の(x1, y1)と右下の(x2, y2)）
x1, y1, x2, y2 = 600, 420, 640, 500

# スナップショットの表示
cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
ret, frame = cap.read()
if ret:
    cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
    plt.figure(figsize=(8, 8))
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.title('Snapshot with ROI')

plt.show()
cap.release()
```

```python
### 動画フレームを読んでROI内の平均輝度を計算し、時系列をプロットする

cap = cv2.VideoCapture(video_path)

# 輝度平均を格納するリストを用意
luminance_means = []

# 動画をフレームごとに処理
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # 矩形領域の抽出
    roi = frame[y1:y2, x1:x2]

    # 輝度（グレースケール）に変換し、平均を計算
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    luminance_mean = np.mean(gray)
    luminance_means.append(luminance_mean)
cap.release()

# 平均輝度の時系列をプロット
plt.figure(figsize=(10, 5))
plt.plot(luminance_means)
plt.title('Luminance Average Over Time')
plt.xlabel('Frame')
plt.ylabel('Average Luminance')
plt.show()
```

```python
### 時系列信号にバンドパスフィルタを適用し、心拍由来と考えられる信号（通常は1-2Hz程度のピーク）が見られるか確認する

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt
from numpy.fft import fft, fftfreq

# バターワース・バンドパスフィルタの設定
nyquist_freq = 0.5 * fps
low_cutoff = 0.2  # バンドパスフィルタの下限周波数
high_cutoff = 6  # バンドパスフィルタの上限周波数
order = 5  # フィルタの次数

# フィルタ係数の計算
b, a = butter(order, [low_cutoff / nyquist_freq, high_cutoff / nyquist_freq], btype='band')

# フィルタの適用
filtered_luminance_means = filtfilt(b, a, luminance_means)

# フーリエ変換の実行
fft_data = fft(filtered_luminance_means)
amplitude_spectrum = np.abs(fft_data)

# 周波数軸の作成
dt = 1 / fps
N = len(filtered_luminance_means)
xf = fftfreq(N, dt)

# プロット
plt.figure(figsize=(10, 10))

plt.subplot(2, 1, 1)
plt.plot(filtered_luminance_means)
plt.title('Filtered Luminance Average Over Time')
plt.xlabel('Frame')
plt.ylabel('Average Luminance')

plt.subplot(2, 1, 2)
plt.plot(xf, amplitude_spectrum)
plt.title('Amplitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.xlim([0, 6])

plt.show()
```

```python
### 短時間フーリエ変換を行い、スペクトログラムを表示する

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft

# 時系列データ
time_series = filtered_luminance_means

# 短時間フーリエ変換の実行
f, t, Zxx = stft(time_series, fs=fps, nperseg=fps*8)

# プロット
plt.figure(figsize=(10, 5))
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.ylim(0,6)
plt.show()
```
